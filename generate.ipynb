```python
%matplotlib inline
from captcha.image import ImageCaptcha 
from PIL import Image
import matplotlib.pyplot as plt
import string

characters = string.digits + string.ascii_uppercase
width, height, n_len, n_class = 160, 60, 4, len(characters)
generator = ImageCaptcha(width=width, height=height)
# generate aplha-type captcha
label = 'abcd'
image = generator.generate_image(label)
plt.imshow(image)
plt.title(label)
```

```python
# generate num-type captcha
label = '1234'
image = generator.generate_image(label)
plt.imshow(image)
plt.title(label)
```

```python
#generate dataset
import random

random.seed(2021)

for i in range(10000):
    random_str = ''.join([random.choice(characters) for _ in range(4)])
    image = generator.generate_image(random_str)
    image.save('images/%s.jpg' % random_str)
```
